## Characters and Character Relationships

```python
%sql use gotprocessed;
```

```python
%%sql characters_core_info << 
        select distinct character_id, characterName from characters join
                character_relationships_fixed as crf
            on 
                characters.character_id = crf.source_character_id;
```

```python
characters_core_info[0]
```

```python
cypher_template = """
    create (:Character { character_id: $character_id,
                            character_name: $character_name
                        }
            )
"""
```

```python
cid = characters_core_info[0][0]
cname = characters_core_info[0][1]

cid, cname
```

```python
graph.run(cypher_template, character_id=cid, character_name=cname)
```

```python
characters_core_info = characters_core_info[1:]
```

```python
for c in characters_core_info:
    cid = c[0]
    cname = c[1]
    graph.run(cypher_template, character_id=cid, character_name=cname)
```

```python
%%sql

characters_relationship_info << 
        select distinct source_character_id, target_id as target_character_id, label as relationship from characters join
                character_relationships_fixed as crf
            on 
                characters.character_id = crf.source_character_id;
```

```python
characters_relationship_info
```

```python
relationship = characters_relationship_info[0]['relationship'].upper()
cypher_template = """
    match 
        (s:Character { character_id: $source_character_id}),
        (t:Character { character_id: $target_character_id})
    create
        (s)-[:""" \
    + \
    relationship \
    + \
    """]->(t)
    """
```

```python
sid = characters_relationship_info[0]['source_character_id']
tid = characters_relationship_info[0]['target_character_id']
```

```python
print(cypher_template)
```

```python
(sid, tid)
```

```python
graph.run(cypher_template, source_character_id=sid, target_character_id=tid)
```

```python
for r in characters_relationship_info:
    
    sid = r['source_character_id']
    tid = r['target_character_id']
    relationship = r['relationship'].upper()
    
    cypher_template = """
    match 
        (s:Character { character_id: $source_character_id}),
        (t:Character { character_id: $target_character_id})
    create
        (s)-[:""" \
    + \
    relationship \
    + \
    """]->(t)
    """
    
    graph.run(cypher_template, source_character_id=sid, target_character_id=tid)
    
```
