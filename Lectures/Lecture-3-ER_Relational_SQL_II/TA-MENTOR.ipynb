```python
%%sql 

select @ta_count:=count(*) from tas;
```

```python
%%sql ta_student << with student_ta as (
	select uni, tas.id as ta_id, a.first_name as student_first_name, a.last_name as student_last_name, 
		tas.name as ta_name,  `E-mail 1 - value` as ta_email from
		(select id as student_id, floor(id/@ta_count)+1 as mentor_id, uni, first_name, last_name from classroster) as a
		join
		tas
		where
		mentor_id=tas.id)
select *   from student_ta;
```
